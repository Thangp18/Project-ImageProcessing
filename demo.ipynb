```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def hieuChinhGamma(image, gamma=1.0):
    """
    Áp dụng phép biến đổi hàm mũ (Hiệu chỉnh Gamma) cho ảnh.
    
    Args:
        image: Ảnh đầu vào (đọc bằng OpenCV).
        gamma: Giá trị gamma. gamma < 1 làm sáng ảnh, gamma > 1 làm tối ảnh.
        
    Returns:
        Ảnh đã được hiệu chỉnh gamma.
    """
    gamma
    reipe = np.array([((i / 255.0) ** gamma) * 255
                      for i in np.arange(0, 256)])
    
    return cv2.LUT(image, reipe)

try:
    image = cv2.imread('ảnh thiếu sáng.jpg')
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
except:
    print("Không tìm thấy ảnh! Tạo một ảnh demo thiếu sáng.")
    gray_image = np.full((300, 400), 50, dtype=np.uint8)
    cv2.putText(gray_image, 'Dark Image', (50, 150), cv2.FONT_HERSHEY_SIMPLEX, 2, (120, 120, 120), 3)
```

```python
image_tangDoSang = hieuChinhGamma(gray_image, 0.5)
image_tangDoToi = hieuChinhGamma(gray_image, 2.0)

# Hiển thị kết quả
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_tangDoSang, cmap='gray')
plt.title(f'Làm sáng (Gamma = {0.5})')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_tangDoToi, cmap='gray')
plt.title(f'Làm tối (Gamma = {2.0})')
plt.axis('off') 

plt.tight_layout()
plt.show()
```
